# <font color= #2E9AFE> Student Performance Factors</font>

El dataset muestra 20 columnas para comprender el rendimieno de los estudiantes en los examenes. Tiene 6,607 registros, para su obtención los pueden consultar en: https://www.kaggle.com/datasets/lainguyn123/student-performance-factors

- **Hours_Studied**: Número de horas dedicadas al estudio por semana.
- **Attendance**: Porcentaje de clases a las que se asistió.
- **Parental_Involvement**: Nivel de participación de los padres en la educación del estudiante (Bajo, Medio, Alto).
- **Access_to_Resources**: Disponibilidad de recursos educativos (Bajo, Medio, Alto).
- **Extracurricular_Activities**: Participación en actividades extracurriculares (Sí, No).
- **Sleep_Hours**: Promedio de horas de sueño por noche.
- **Previous_Scores**: Calificaciones de exámenes anteriores.
- **Motivation_Level**: Nivel de motivación del estudiante (Bajo, Medio, Alto).
- **Internet_Access**: Disponibilidad de acceso a internet (Sí, No).
- **Tutoring_Sessions**: Número de sesiones de tutoría a las que se asistió por mes.
- **Family_Income**: Nivel de ingresos familiares (Bajo, Medio, Alto).
- **Teacher_Quality**: Calidad de los profesores (Bajo, Medio, Alto).
- **School_Type**: Tipo de escuela (Pública, Privada)
- **Peer_Influence**: Influencia de los compañeros (Positiva, Neutral, Negativa)
- **Physical_Activity**: Actividad física (horas por semana)
- **Learning_Disabilities**: Discapacidades de aprendizaje (Sí, No)
- **Parental_Education_Level**: Nivel educativo de los padres (Secundaria, Universidad, Posgrado)
- **Distance_from_Home**: Distancia al hogar (Cerca, Moderada, Lejana)
- **Gender**: Género (Masculino, Femenino)

**Target o variable a predecir**
- Exam_Score: Calificación final del examen
